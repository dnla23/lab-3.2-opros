```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_excel('DE .xlsx')
```

```python
df
```

Оценка осведомленности в области Data Engineering

```python
plt.figure(figsize=(8, 6))
sns.histplot(df["Как вы оцениваете свою осведомленность в области Data Engineering?"], bins=5, kde=True)
plt.title("Оценка осведомленности в области Data Engineering")
plt.xlabel("Оценка")
plt.ylabel("Количество респондентов")
plt.show()
```

Использование инструментов

```python
tools = df["Какие из следующих инструментов вы используете в своей работе?"].str.split(", ", expand=True).stack().value_counts()
plt.figure(figsize=(10, 6))
tools.plot(kind='bar')
plt.title("Использование инструментов в работе")
plt.xlabel("Инструменты")
plt.ylabel("Количество упоминаний")
plt.show()
```

Важность SQL для Data Engineer

```python
sql = df["Как вы оцениваете важность SQL для Data Engineer?"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(sql, labels=sql.index, autopct='%1.1f%%', startangle=140)
plt.title("Важность SQL для Data Engineer")
plt.show()
```
